```python
def doi_process(d):
    if d.startswith('http://'):
        return d
    elif d.startswith('doi.org'):
        return f'http://{d}'
    else: 
        return f'http://doi.org/{d}'
    
```

```python
metadata['doi'] = metadata['doi'].fillna(' ').apply(doi_process)
```
